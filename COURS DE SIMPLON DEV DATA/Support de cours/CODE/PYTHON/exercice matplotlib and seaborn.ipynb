```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("//home//virus//Téléchargements//120-years-of-olympic-history-athletes-and-results1.csv",index_col=[0])

plt.figure()
plt.subplot(111)
plt.plot(data['Height'],
         data['Weight'])

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot()
plt.plot(data["Height"],data["Weight"],marker="o",color="red",linestyle='')

plt.title('Poids des athlètes en fonction de leur taille')
plt.ylabel("Poids")
plt.xlabel("Taille")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[5,5],dpi=150, facecolor='grey')
plt.subplot(111)
plt.plot(data["Height"],data["Weight"],marker="o",color="red",linestyle='',markersize=1.8)

plt.title('Poids en fonction de la taille des athlètes des JOs')
plt.ylabel("Poids")
plt.xlabel("Taille")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(111)
plt.plot(data["Height"],data["Weight"],marker="o",color="red",linestyle='',markersize=2.5)
plt.plot(data["Height"],data["Age"],marker="o",color="blue",linestyle='',markersize=2.5)



plt.title('Poids en fonction de la taille des athlètes (rouge)  \
et Age en fonction de la taille des athlètes (bleu)')
plt.ylabel("Poids/Age")
plt.xlabel("Taille")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(121)
plt.plot(data["Height"],data["Weight"],marker="o",color="red",linestyle='',markersize=2.5)
plt.title('Athlètes des JOs')
plt.ylabel("Poids")
plt.xlabel("Taille")

plt.subplot(122)

plt.plot(data["Height"],data["Age"],marker="o",color="blue",linestyle='',markersize=2.5)



plt.title('Athlètes des JOs')
plt.ylabel("Age")
plt.xlabel("Taille")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(111)
plt.plot(data["Height"],data["Weight"],marker="o",color="red",linestyle='',markersize=2.5, label='Poids_vs_Taille')
plt.plot(data["Height"],data["Age"],marker="v",color="blue",linestyle='',markersize=2.5,label='Age_vs_Taille')


plt.title('Athlètes des JOs')
plt.ylabel("Poids/Age")
plt.xlabel("Taille")
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(111)
plt.plot(data["Height"],data["Weight"],marker="o",color="red",linestyle='',markersize=2.5, label='Poids_vs_Taille')
plt.plot(data["Height"],data["Age"],marker="v",color="blue",linestyle='',markersize=2.5,label='Age_vs_Taille')

plt.annotate("Moyenne d'age = 25 ans", xy=(210,25), xytext=(250,200), arrowprops=dict(facecolor='black'))

plt.title('Athlètes des JOs')
plt.ylabel("Poids/Age")
plt.xlabel("Taille")
plt.legend(loc='upper left')
plt.show()
```
