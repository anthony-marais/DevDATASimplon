```python
# Import de données depuis un fichié Excel dans un dataframe

nom_colonnes_p_emploi = ['code', 'Libellé', 'Pôle emploi 2018']

p_emploi = pd.read_excel("//home//virus//Documents//Analyse exploratoire taux de chômage//pole_emploi.xlsx",
                     names=nom_colonnes_p_emploi, sheet_name='Data', skiprows=4)

# Afficher les premières lignes du dataframe

(p_emploi)
```

```python
#data.merge(p_emploi, left_on='lkey', right_on='rkey')

display(data,p_emploi)
```

```python

# • Joindre les data et p_emploi au seins d’un seul et unique dataframe en faisant 
#correspondre les régions par leur codes. Utiliser pd.merge() pour réaliser cette jointure.

full_data = pd.merge(data, p_emploi, left_on='REG', right_on='code')


(full_data)
```

```python
#• Créer une colonne quartile dans p_emploi qui prendra les valeurs [‘q25’, ‘q50’, ‘q75’, 
#‘q100’] en fonction du quartile auquel appartient le nombre de pôle emploi. Pour se faire 
#utiliser « pd.qcut() »

p_emploi['quartile'] = pd.qcut(p_emploi['Pôle emploi 2018'], q = 4, labels = ['q25', 'q50', 'q75', 'q100'])


p_emploi.sort_values(by=['quartile'])
    


```

```python
#• Générer un pivot de table montrant la moyenne du taux de chômage du trimestre 1 de 
# l’année 2020 pour chaque régions et quantile du nombre de pôle emploi.
#• Appeler la méthode pivot_table () sur le data frame
#• Passer au paramètre « index » la valeur « LIBREG ».
#• Passer au paramètre « columns » la valeur « quantile ».
#• Passer au paramètre « values » la valeur «2020-T1 »
#• Afficher les résultats.
```

```python
moyenne_chomage_2020_T1 = pd.pivot_table(full_data, values=['2020-T1'], index = ['LIBREG'], columns = ['quartile'])

moyenne_chomage_2020_T1
```

```python
moyenne_chomage_2019_T1 = pd.pivot_table(full_data, values=['2019-T1'], index = ['LIBREG'], columns = ['quartile'])


moyenne_chomage_2019_T1
```

```python
moyenne_chomage_2018_T1 = pd.pivot_table(full_data, values=['2018-T1'], index = ['LIBREG'], columns = ['quartile'])

moyenne_chomage_2018_T1
```
