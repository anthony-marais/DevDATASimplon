```python
# Jeu de monstre en objet

import random
from time import sleep

class case:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def adjacentes(self, partie):
        for c in partie.liste_des_cases:
            return [c for c in partie.liste_des_cases if self != c and abs(c.x-self.x)<2 and abs(c.y-self.y)<2]
    def __str__(self):
        return "x:{}, y:{}".format(self.x, self.y)
        
class creature:
    def __init__(self, nom, position):
        self.nom = nom
        self.position = position
    def choisirCible(self, partie):
        cases_jouable = partie.actif.position.adjacentes(partie)
        partie.actif.position = next((case_cible for case_cible in cases_jouable if partie.est_occupe(case_cible) == True), random.choice(cases_jouable))
        print("{} se déplace à la position => {}".format(partie.actif.nom, partie.actif.position))
        
class jeu:
    def __init__(self, liste_des_cases, liste_des_creatures):
        self.liste_des_cases = [case(x, y) for x in [i for i in range(1, liste_des_cases[0] + 1)] for y in [i for i in range(1, liste_des_cases[1]  + 1)]]
        self.liste_des_creatures = [creature(nom_creature, self.liste_des_cases[index if index == 0 else -1]) for index, nom_creature in enumerate(liste_des_creatures)]
        self.tours = 1
        self.actif = random.choice(self.liste_des_creatures)
    def afficher_jeu(self):
        lignes = [['']*(max(case_x_max.x for case_x_max in self.liste_des_cases)) for x in range(max(case_y_max.y for case_y_max in self.liste_des_cases))]
        for monstre in self.liste_des_creatures:
            lignes[monstre.position.y-1][monstre.position.x-1] = (self.actif.nom if lignes[monstre.position.y-1][monstre.position.x-1] !="" else monstre.nom)
        largeur_colonnes = [max(len(x.nom) for x in self.liste_des_creatures)]*(max(case_x_max.x for case_x_max in self.liste_des_cases))
        for colonnes in reversed(lignes):
            colonne = [str(c).ljust(w) for w, c in zip(largeur_colonnes, colonnes)]
            print("|{}|".format("|".join(list(colonne))))
            print(''.ljust(largeur_colonnes[0]*len(largeur_colonnes)+(len(largeur_colonnes)+1), '-'))
    def est_occupe(self, position):
         return next((True for monstre in self.liste_des_creatures if monstre.nom != self.actif.nom and monstre.position == position), False)
    def deplacer(self):
        print("C'est l'heure du du du duel !\n")
        print("            |")
        print("////////////|---------------------------------,")
        print("`^^^^^^^^^^^|--------------------------------\"")
        print("            |")
        print("\nC'est {} qui engage à la position => {}".format(self.actif.nom, self.actif.position))
        self.afficher_jeu()
        print("\nTour: ", self.tours)
        self.actif.choisirCible(self)
        while self.est_occupe(self.actif.position) != True:
            self.afficher_jeu()
            self.actif = self.liste_des_creatures[0 if self.liste_des_creatures.index(self.actif) == len(self.liste_des_creatures)-1 else self.liste_des_creatures.index(self.actif)+1]
            self.tours += 1
            print("\nTour: ", self.tours)
            print("C'est au tour de: ", self.actif.nom, flush=True)
            sleep(random.randint(2,4))
            self.actif.choisirCible(self)
        self.afficher_jeu()
        print("\nIncroyable! {} l'emporte en {} Tours!".format(self.actif.nom, self.tours))
            
    
partie = jeu([4, 4], ['Rick', 'Morty'])
partie.deplacer()
```
