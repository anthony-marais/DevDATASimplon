```python

# Import de la librairie pandas avec l'allias pd

import pandas as pd

# importation des données du fichier excel

data = pd.read_excel("//home//virus//Documents//Data visualisation du taux de chômage//chomage_zone.xls",
                     sheet_name='txcho_ze' ,skiprows=5,)

data
```

```python
#• En utilisant groupby() et mean() successivement calculer le taux de 
# chômage moyen par régions pour chaque trimestres. Dans le groupby
# mettre as_index=‘False’. L’assigner à data_grp



data_grp = (data.groupby('LIBREG' ,as_index=False).mean())

data_grp
```

```python
# Utiliser pandas.plot() pour tracer l’diagramme à barre du taux de chômage du 
# premier trimestre 2020. Les noms de régions doivent apparaitre en abscisse
# En vous appuyant sur la documentation en ligne donné une taille correct 
# au graphique, lui donner un titre et ajouter une grille en arrière-plan.



t1_2020_diagramme_barre = (data_grp.plot(x='LIBREG',y = '2020-T1',kind = 'bar' ,figsize = (20,10), grid=True,sharex=True,sharey=True,title='taux de chômage du premier trimestre 2020'))


#print(t1_2020_diagramme_barre.text(10.10, 10.10, 'taux de chômage du premier trimestre 2020',horizontalalignment='center',verticalalignment='center'))
```

```python
# • En repartant du jeux de données originel, isoler les colonnes 3 à 73. et assignées les à data2.


data2 = data.iloc[:,3:73]

data2
```

```python
# En utilisant groupby() et mean() successivement calculer le taux de chômage moyen par 
# régions pour chaque trimestres. Cette fois ci ne pas préciser l’instruction as_index. L’assigner à data_grp2.

data_grp2 = data2.groupby('LIBREG').mean()

data_grp2
```

```python
# • Transposer le dataframe

data_transpose = data_grp2.transpose()

data_transpose
```

```python
# • Réindexer le dataframe avec resetindex

data_transpose = data_transpose.reset_index()

data_transpose
```

```python
# • Renommer le nom de la colonne ‘index’ en ‘periode

data_transpose = data_transpose.rename(columns={"index": "periode"})

data_transpose
```

```python
# • Après avoir visualisé le contenu de data2, utiliser pandas.plot() pour tracer la ligne 
# représentant l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte d'Azur

data_transpose.plot(x = "periode",y = "Provence-Alpes-Côte d'Azur" , figsize = (20,10))
```

```python
#  Répéter l’étape précédente sans préciser l’argument « y = »

data_transpose.plot(x = "periode", figsize = (20,10))
```

```python
# • Représenter l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte d'Azur’,
# l‘Occitanie’ et ‘Ile-de-France’ sur un même graphique.

data_transpose.plot(x = "periode",y =["Provence-Alpes-Côte d'Azur", "Occitanie", "Ile-de-France"  ], figsize = (20,10))
```
