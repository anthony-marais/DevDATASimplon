```python
#18. Tracer un graphique représentant le nombre de séries par modalité de nombre de saisons

data_serie_2 = pd.DataFrame(data_2.loc[data['type']=="TV Show"])
data_serie_2.head(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
plt.figure(figsize=[20,10])
plt.subplot(111)
plt.title('Nombre de films/séries par classement de contenu', fontsize="20")
ab = sns.countplot(x = "duration", data=data_serie_2, order=data_serie_2['duration'].value_counts().keys())
plt.xlabel('Durée en saison', fontsize="15")
plt.ylabel('Nombre de Series', fontsize="15")
plt.show()
```
